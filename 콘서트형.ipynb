**방문 도시와 관객수의 상관관계**

```python
#전체 아이돌 정보
total_city_seats = pd.DataFrame()

for i in range( len(world_tour) ):
  df = world_tour[i]
  tour_names = df.iloc[1:, 0].dropna().unique().tolist()

  # [투어명, 방문도시수, 평균관객수] 데이터프레임 생성
  city_seats = pd.DataFrame( tour_names, columns = ["투어명"])

  visiting_city = [] #방문도시수
  seats = [] #평균관객수

  for name in tour_names:
      tour = df.loc[df.iloc[:, 0] == name, :]

      # 방문 도시 수 계산
      city_count = tour.iloc[:, 3].nunique()
      visiting_city.append(city_count)

      # ~명, 10,000 등을 숫자로 바꾸고 결측치 제거
      seat = tour.iloc[:, 5].apply(lambda x: x.replace('명', '').replace(',', '') if isinstance(x, str) else x).dropna()
      seat = seat.astype(float)  # 숫자로 변환
      seat_mean = round(np.mean(seat), 2)
      seats.append(seat_mean)

  city_seats["방문 도시 수"] = visiting_city
  city_seats['평균 관객 수'] = seats

  total_city_seats = pd.concat([total_city_seats, city_seats], ignore_index=True)
```
